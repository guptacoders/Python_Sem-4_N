```python
import pandas as pd
df = pd.read_csv("Police_Department_Incidents_-_Previous_Year__2016_.csv")
df.head()
```

```python
df.shape
```

```python
df=df.head(100)
df.shape
```

```python
latitude = 37.77
longitude = -122.42
sanfran_map = folium.Map(location=[latitude,longitude],zoom_start=12)
sanfran_map
```

```python
incidents = folium.map.FeatureGroup()
for lat,lng in zip(df.Y,df.X):
    incidents.add_child(folium.features.CircleMarker([lat,lng],radius=6,color="red",fill=True,fill_color="white",fill_opacity=0.7))
sanfran_map.add_child(incidents)
```

```python
# labels=[df.Category]
for lat,lng,label in zip(df.Y,df.X,df.Category):
    folium.Marker([lat,lng],popup=label).add_to(sanfran_map)
    
sanfran_map.add_child(incidents)
```

# CHROPLETH MAPS

```python
import pandas as pd
import folium
state_unemp = pd.read_csv("US_Unemployment_Oct2012.csv")
state_unemp.head()
```

```python
state_geo="us-states.json"
usa_state=folium.Map(location=[48,-102],zoom_start=3)
folium.Choropleth(geo_data=state_geo,name="choropleth",data=state_unemp,columns=["State","Unemployment"],key_on="feature.id",fill_colour="YlGnBu",fill_opacity=0.7,line_opacity=0.2,legend_name="Unemployement Scale").add_to(usa_state)
usa_state
```

# NetWorkx Library

```python
import networkx as nx
import matplotlib.pyplot as plt

# Undirected Graph

G = nx.Graph()
G.add_node(1)
G.add_nodes_from([2,3,4])
G.add_nodes_from(range(5,11))
G.add_edge(1,2)
G.add_edges_from([(2,3),(3,4),(4,5),(5,6),(6,8),(8,7),(7,1),(1,10),(1,9)])
plt.figure(figsize=(10,10))


nx.draw_networkx(G,node_size=850,node_color="green",edge_color="red",width=5)
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Directed Graph

G = nx.DiGraph()
G.add_node(1)
G.add_nodes_from([2,3,4])
G.add_nodes_from(range(5,11))
G.add_edge(1,2)
G.add_edges_from([(2,3),(3,4),(4,5),(5,6),(6,8),(8,7),(7,1),(1,10),(1,9)])
plt.figure(figsize=(10,10))


nx.draw_networkx(G,node_size=850,node_color="green",edge_color="red",width=5)
plt.show()
```

```python
# QB-87
# You have been hired by an Airlines company to analyze their routes. The company has provided you following data.
# Your task is to create a NetworkX directed graph representing the routes and display it.
# Figure size should be (15,15), node color should be green, take appropriate node size, edge color should be red.
# Data:
# Kolkata to Mumbai 
# Mumbai to Pune 
# Mumbai to Goa
# Kolkata to Delhi 
# Kolkata to Bhubaneshwar 
# Mumbai to Delhi 
# Delhi to Chandigarh 
# Delhi to Surat 
# Kolkata to Hyderabad 
# Hyderabad to Chennai 
# Chennai to Thiruvananthapuram 
# Thiruvananthapuram to Hyderabad 
# Kolkata to Varanasi 
# Delhi to Varanasi 
# Mumbai to Bangalore 
# Chennai to Bangalore 
# Hyderabad to Bangalore 
# Kolkata to Guwahati 
import networkx as nx
import matplotlib.pyplot as plt


L = ["Kolkata","Mumbai","Pune","Goa","Kolkata","Delhi","Bhubaneshwar","Chandigarh","Surat","Hyderabad","Chennai","Thiruvananthapuram","Varanasi","Bangalore","Guwahati"]
N =nx.DiGraph()
plt.figure(figsize=(10,10))

N.add_edges_from([(L[0],L[1]),(L[1],L[2]),(L[1],L[3]),(L[0],L[5]),(L[0],L[6]),(L[1],L[5]),(L[5],L[7]),(L[5],L[8]),(L[0],L[9]),(L[0],L[5])])
nx.draw_networkx(N,node_size=2500,node_color="green",edge_color="red",width=3)
plt.show()
```

```python
# QB_72
# The following dictionary shows how five people follow each other on Instagram:
#  instagram = {'person1': [0,1,1,0,1], 'person2': [0,0,1,0,1], 'person3': [1,1,0,1,1],'person4': [1,1,1,0,0], 'person5': [1,1,0,0,0]}
#  E.g., the list for person1 has the value on index 2 as 1 which means person1 followsperson3 and a directed edge should be 
# added from person1 to person3.
#  Using networkx library, create a directed graph.

import networkx as nx
import matplotlib.pyplot as plt

G=nx.DiGraph()
plt.figure(figsize=(10,10))

instagram = {'person1': [0,1,1,0,1], 'person2': [0,0,1,0,1], 'person3': [1,1,0,1,1],'person4': [1,1,1,0,0], 'person5': [1,1,0,0,0]}
K=list(instagram.keys())
V=list(instagram.values())
G.add_nodes_from(K)

for i in range(0,5):
    for j in range(0,5):
        if V[i][j]==1:
            G.add_edge(K[i],K[j])
            
nx.draw_networkx(G,node_size=2500,node_color="yellow",edge_color="red",width=3)
plt.show()
            
```
