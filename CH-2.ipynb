# Data Visualization (7%)

```python
from wordcloud import WordCloud , STOPWORDS

alice_novel = open("alice.txt",encoding="utf-8")
data = alice_novel.read()
print(data)
```

```python
stop = set(STOPWORDS)
stop
```

```python
alice_wl = WordCloud(background_color="white",max_words=2000,stopwords=stop)
```

```python
alice_wl.generate(data)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(14,18))
plt.imshow(alice_wl)
plt.axis("off")
plt.show()
```

```python
stop.add("said")
```

```python
alice_wl.generate(data)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(14,18))
plt.imshow(alice_wl)
plt.axis("off")
plt.show()
```

# Waffle Chart

```python
import pandas as pd
from pywaffle import Waffle

data = {"Phone":["Xiomi","Samsung","Apple","Nokia","Realme"],
         "Stock":[44,12,8,5,50]}
df=pd.DataFrame(data)
df
```

```python
fig = plt.figure(FigureClass=Waffle,rows=15,values=df["Stock"],labels=list(df["Phone"]))
plt.show()
```
