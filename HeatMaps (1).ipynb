```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
D=np.random.randint(low=1,high=100,size=(10,10))
print(D)
```

```python
hm=sns.heatmap(data=D)
plt.show()
```

```python
D=np.random.randint(low=1,high=100,size=(10,10))
# print(D)
hm=sns.heatmap(data=D,vmax=70,vmin=30)
plt.show()
```

```python
D=np.random.randint(low=1,high=100,size=(10,10))
hm=sns.heatmap(data=D,cmap="tab20")
plt.show()
```

```python
D=np.random.randint(low=1,high=100,size=(10,10))
# plt.figure(figsize=(10,10))
hm=sns.heatmap(data=D,cmap="tab20",annot=True)
plt.show()
```

```python
D=np.random.randint(low=1,high=100,size=(10,10))
hm=sns.heatmap(data=D,cmap="tab20",annot=True,cbar=False,xticklabels=False,yticklabels=False,linewidths=3,linecolor="black")
plt.show()
```

```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/kavit88/Data-Sets/main/restaurant_reviews.csv")
df
```

```python
s=df["review_text"]
print(s)
```

```python
x=""
for i in s:
    x+=i
    
print(x)
```

```python
from wordcloud import WordCloud , STOPWORDS
stop = set(STOPWORDS)
obj = WordCloud(background_color="white",max_words=150,stopwords=stop)
obj.generate(x)
plt.figure(figsize=(14,18))
plt.imshow(obj)
plt.axis("off")
plt.show()
```
