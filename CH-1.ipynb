# 1) Pandas Series

Pandas Series Create Using by List.

```python
import pandas as pd
a = [1,7,2]
s = pd.Series(a)
print(s)
print(s[1])  # Index Data Fetching
```

```python
a = [1,7,2]
s = pd.Series(a,index=["x","y","z"])
print(s)
print(s["z"])
```

Pandas Series Create Using by Dictionary...

```python
import pandas as pd
calories = {"day1":420,"day2":380,"day3":390}
s=pd.Series(calories)
print(s)
print("-"*50)
s=pd.Series(calories,index=["day1","day2"])
print(s)
```

# Data Cleaning

```python
import numpy as np
import pandas as pd

sales_data = {"Name":["willium","emma","sofia","markus","edwards","thomas","ethan",np.nan,"arun","ankita","paulo"],
               "Region":[np.nan,"North","East",np.nan,"West","West","South",np.nan,"West","East","South"],
               "Sales":[50000,52000,np.nan,np.nan,42000,72000,49000,np.nan,67000,65000,67000],
               "Expenses":[42000,43000,np.nan,np.nan,38000,39000,42000,np.nan,39000,50000,45000]}

df = pd.DataFrame(sales_data)
df
```

```python
df.isna().sum()  # Null Value Colum Wise With Total
```

```python
# df.dropna()  ---> This function remove null value
# df.fillna()  ---> This function filling to null value
```

```python
df.dropna(thresh=2)  # evi row ne delete karse jema atlest 2 valid data nai hoy (2 valid data hova joiye)
```

```python
# remove all rows with any number of nulls
df.dropna()
```

```python
# Remove Only if All Values Are Null
df.dropna(how="all")
```

```python
# drop only if sales or expences are null
df.dropna(subset=["Sales","Expenses"])
```

```python
# remove all rows then occur null value
df.dropna(axis=0)
```

```python
# remove all Column then occur null value
df.dropna(axis=1)
```

```python
df["Sales"].fillna(0)
```

```python
df["Sales"].fillna(df["Sales"].mean())
# all null value in Sales Coulm replase with average value
```

```python
df["Sales"].fillna(df["Sales"].median())
```

# FINDING OUTLIERS

```python
# FINDING OUTLIERS
import pandas as pd
df = pd.read_csv("auto-mpg.csv")
df
```

```python
# MOST IMP PROGRAM
Q1 = df["acceleration"].quantile(0.25)
Q3 = df["acceleration"].quantile(0.75)
IQR = Q3-Q1
Low = Q1 - (1.5 * IQR)
High = Q3 + (1.5 * IQR)
print("Lower:",Low)
print("Higher:",High)

df= df[(df["acceleration"] < Low) | (df["acceleration"] > High)]
df
```

# REMOVE OUTLIERS

```python
import pandas as pd
df = pd.read_csv("auto-mpg.csv")

Q1 = df["acceleration"].quantile(0.25)
Q3 = df["acceleration"].quantile(0.75)
IQR = Q3-Q1
Low = Q1 - (1.5 * IQR)
High = Q3 + (1.5 * IQR)
print("Lower:",Low)
print("Higher:",High)

df= df[(df["acceleration"] >= Low) & (df["acceleration"] <= High)]
df
```
